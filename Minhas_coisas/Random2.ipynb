```python
import random

def jogo_adivinhe_cor():
    # Lista expandida de cores
    cores = [
        "vermelho", "azul", "verde", "amarelo", "laranja", "roxo", "rosa", 
        "preto", "branco", "cinza", "violeta", "turquesa", "salmon", "bege", 
        "marrom", "magenta", "ciano", "pêssego", "chartreuse", "lavanda"
    ]
    
    # Definindo o número máximo de tentativas
    max_tentativas = 5
    
    # Escolhendo uma cor aleatória
    cor_aleatoria = random.choice(cores)
    
    # Função para dar pistas
    def dar_pista(cor):
        if cor in ["vermelho", "laranja", "amarelo"]:
            return "Dica: É uma cor quente!"
        elif cor in ["azul", "verde", "roxo"]:
            return "Dica: É uma cor fria!"
        elif cor in ["preto", "branco", "cinza", "bege", "marrom"]:
            return "Dica: É uma cor neutra!"
        else:
            return "Dica: É uma cor vibrante e intensa!"

    print("Bem-vindo ao Jogo 'Adivinha a Cor'!")
    print(f"Tente adivinhar a cor que estou pensando. Você tem {max_tentativas} tentativas.")
    
    tentativas = 0
    acertou = False
    
    while not acertou and tentativas < max_tentativas:
        tentativas += 1
        palpite = input(f"Tentativa {tentativas}/{max_tentativas}: Qual é a cor? ").lower()

        # Opção de sair do jogo
        if palpite == "sair":
            print("Você decidiu sair do jogo. Até a próxima!")
            break

        if palpite == cor_aleatoria:
            acertou = True
            print(f"Parabéns! Você adivinhou a cor {cor_aleatoria} em {tentativas} tentativas!")
        else:
            # Dar dica se o jogador errar
            print(f"Errado! {dar_pista(cor_aleatoria)}")
            if tentativas < max_tentativas:
                print(f"Tente novamente!")
            else:
                print(f"Você não adivinhou a cor. A cor era {cor_aleatoria}.")
    
    if acertou:
        print("Você ganhou! 🎉")
    else:
        print("Fim de jogo! Você não conseguiu adivinhar a cor.")

# Iniciar o jogo
jogo_adivinhe_cor()
```

```python
print("=== SELECIONE O QUE VOCÊ COMEU NO CAFÉ DA MANHÃ ===")

# Dicionário com os alimentos e seus valores nutricionais
menu = {
    "Café com açúcar": {"calorias": 76, "carboidratos": 19, "proteinas": 0.2, "gorduras": 0},
    "Iogurte": {"calorias": 71, "carboidratos": 13, "proteinas": 1.7, "gorduras": 1.3},
    "Banana": {"calorias": 40, "carboidratos": 10.32, "proteinas": 0.52, "gorduras": 0.12},
    "Pão de queijo (5 unidades)": {"calorias": 355, "carboidratos": 62.60, "proteinas": 11, "gorduras": 6.20}
}

# Exibir o menu
print("Menu do Café da Manhã:")
for i, item in enumerate(menu.keys(), 1):
    print(f"{i}. {item}")
    
# Captura a escolha do usuário
escolhas = input("Digite os números dos itens que você comeu no café da manhã, separados por vírgula: ")
escolhas = [int(num.strip()) for num in escolhas.split(',') if num.strip().isdigit()]

# Cálculo dos totais usando list comprehension
calorias_totais, carboidratos_totais, proteinas_totais, gorduras_totais = [
    sum(menu[list(menu.keys())[i-1]][nutriente] for i in escolhas if 1 <= i <= len(menu))
    for nutriente in ["calorias", "carboidratos", "proteinas", "gorduras"]
]

# Exibir o total nutricional
print("\n=== TOTAL NUTRICIONAL DO CAFÉ DA MANHÃ ===")
print(f"Calorias totais: {calorias_totais} kcal")
print(f"Carboidratos totais: {carboidratos_totais:.2f} g")
print(f"Proteínas totais: {proteinas_totais:.2f} g")
print(f"Gorduras totais: {gorduras_totais:.2f} g")
```
