```python
import random

def jogo_adivinhe_cor():
    # Lista expandida de cores
    cores = [
        "vermelho", "azul", "verde", "amarelo", "laranja", "roxo", "rosa", 
        "preto", "branco", "cinza", "violeta", "turquesa", "salmon", "bege", 
        "marrom", "magenta", "ciano", "p√™ssego", "chartreuse", "lavanda"
    ]
    
    # Definindo o n√∫mero m√°ximo de tentativas
    max_tentativas = 5
    
    # Escolhendo uma cor aleat√≥ria
    cor_aleatoria = random.choice(cores)
    
    # Fun√ß√£o para dar pistas
    def dar_pista(cor):
        if cor in ["vermelho", "laranja", "amarelo"]:
            return "Dica: √â uma cor quente!"
        elif cor in ["azul", "verde", "roxo"]:
            return "Dica: √â uma cor fria!"
        elif cor in ["preto", "branco", "cinza", "bege", "marrom"]:
            return "Dica: √â uma cor neutra!"
        else:
            return "Dica: √â uma cor vibrante e intensa!"

    print("Bem-vindo ao Jogo 'Adivinha a Cor'!")
    print(f"Tente adivinhar a cor que estou pensando. Voc√™ tem {max_tentativas} tentativas.")
    
    tentativas = 0
    acertou = False
    
    while not acertou and tentativas < max_tentativas:
        tentativas += 1
        palpite = input(f"Tentativa {tentativas}/{max_tentativas}: Qual √© a cor? ").lower()

        # Op√ß√£o de sair do jogo
        if palpite == "sair":
            print("Voc√™ decidiu sair do jogo. At√© a pr√≥xima!")
            break

        if palpite == cor_aleatoria:
            acertou = True
            print(f"Parab√©ns! Voc√™ adivinhou a cor {cor_aleatoria} em {tentativas} tentativas!")
        else:
            # Dar dica se o jogador errar
            print(f"Errado! {dar_pista(cor_aleatoria)}")
            if tentativas < max_tentativas:
                print(f"Tente novamente!")
            else:
                print(f"Voc√™ n√£o adivinhou a cor. A cor era {cor_aleatoria}.")
    
    if acertou:
        print("Voc√™ ganhou! üéâ")
    else:
        print("Fim de jogo! Voc√™ n√£o conseguiu adivinhar a cor.")

# Iniciar o jogo
jogo_adivinhe_cor()
```

```python
print("=== SELECIONE O QUE VOC√ä COMEU NO CAF√â DA MANH√É ===")

# Dicion√°rio com os alimentos e seus valores nutricionais
menu = {
    "Caf√© com a√ß√∫car": {"calorias": 76, "carboidratos": 19, "proteinas": 0.2, "gorduras": 0},
    "Iogurte": {"calorias": 71, "carboidratos": 13, "proteinas": 1.7, "gorduras": 1.3},
    "Banana": {"calorias": 40, "carboidratos": 10.32, "proteinas": 0.52, "gorduras": 0.12},
    "P√£o de queijo (5 unidades)": {"calorias": 355, "carboidratos": 62.60, "proteinas": 11, "gorduras": 6.20}
}

# Exibir o menu
print("Menu do Caf√© da Manh√£:")
for i, item in enumerate(menu.keys(), 1):
    print(f"{i}. {item}")
    
# Captura a escolha do usu√°rio
escolhas = input("Digite os n√∫meros dos itens que voc√™ comeu no caf√© da manh√£, separados por v√≠rgula: ")
escolhas = [int(num.strip()) for num in escolhas.split(',') if num.strip().isdigit()]

# C√°lculo dos totais usando list comprehension
calorias_totais, carboidratos_totais, proteinas_totais, gorduras_totais = [
    sum(menu[list(menu.keys())[i-1]][nutriente] for i in escolhas if 1 <= i <= len(menu))
    for nutriente in ["calorias", "carboidratos", "proteinas", "gorduras"]
]

# Exibir o total nutricional
print("\n=== TOTAL NUTRICIONAL DO CAF√â DA MANH√É ===")
print(f"Calorias totais: {calorias_totais} kcal")
print(f"Carboidratos totais: {carboidratos_totais:.2f} g")
print(f"Prote√≠nas totais: {proteinas_totais:.2f} g")
print(f"Gorduras totais: {gorduras_totais:.2f} g")
```
