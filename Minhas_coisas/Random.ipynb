```python
print("=== SELECIONE O QUE VOCÊ COMEU NO CAFÉ DA MANHÃ ===")

# Dicionário com os alimentos e seus valores nutricionais
menu = {
    "Café com açúcar": {"calorias": 76, "carboidratos": 19, "proteinas": 0.2, "gorduras": 0},
    "Iogurte": {"calorias": 71, "carboidratos": 13, "proteinas": 1.7, "gorduras": 1.3},
    "Banana": {"calorias": 40, "carboidratos": 10.32, "proteinas": 0.52, "gorduras": 0.12},
    "Pão de queijo (5 unidades)": {"calorias": 355, "carboidratos": 62.60, "proteinas": 11, "gorduras": 6.20}
}

# Exibir o menu
print("Menu do Café da Manhã:")
for i, item in enumerate(menu.keys(), 1):
    print(f"{i}. {item}")
    
# Captura a escolha do usuário
escolhas = input("Digite os números dos itens que você comeu no café da manhã, separados por vírgula: ")
escolhas = [int(num.strip()) for num in escolhas.split(',') if num.strip().isdigit()]

# Cálculo dos totais usando list comprehension
calorias_totais, carboidratos_totais, proteinas_totais, gorduras_totais = [
    sum(menu[list(menu.keys())[i-1]][nutriente] for i in escolhas if 1 <= i <= len(menu))
    for nutriente in ["calorias", "carboidratos", "proteinas", "gorduras"]
]

# Exibir o total nutricional
print("\n=== TOTAL NUTRICIONAL DO CAFÉ DA MANHÃ ===")
print(f"Calorias totais: {calorias_totais} kcal")
print(f"Carboidratos totais: {carboidratos_totais:.2f} g")
print(f"Proteínas totais: {proteinas_totais:.2f} g")
print(f"Gorduras totais: {gorduras_totais:.2f} g")
```

```python
def calcular_calorias():
    print("🎉 Bem-vindo à Calculadora Mágica de Calorias Diárias! 🍎")
    
    # Solicitar dados do usuário
    print("\nVamos começar! Preencha os seguintes detalhes para descobrir quantas calorias você pode consumir por dia.")
    sexo = input("Qual o seu sexo? (homem/mulher): ").strip().lower()
    
    if sexo not in ['homem', 'mulher']:
        print("Ops! Não entendi... por favor, insira 'homem' ou 'mulher'. Vamos tentar de novo!")
        return
    
    peso = float(input("Qual o seu peso em quilos? ⚖️: "))
    altura = float(input("E qual a sua altura em centímetros? 📏: "))
    idade = int(input("Agora, me diga sua idade em anos! 🎂: "))
    
    print("\nCalculando sua quantidade mágica de calorias... 🔮✨")
    
    if sexo == "homem":
        # Fórmula para homens
        calorias = (13.75 * peso) + (5 * altura) - (6.76 * idade) + 66.5
        print(f"\n💪 Uau! Para um homem com essas características, você pode consumir até {calorias:.2f} calorias por dia! 🥳")
    elif sexo == "mulher":
        # Fórmula para mulheres
        calorias = (9.56 * peso) + (1.85 * altura) - (4.68 * idade) + 665
        print(f"\n💃 Para uma mulher com essas características, você pode consumir até {calorias:.2f} calorias por dia! 💖")
    
    print("\nAgora que você sabe, use essa informação com sabedoria! 😌✨")
    print("Lembre-se: o equilíbrio é a chave para uma vida saudável! 🧘‍♀️🍏")

# Chamar a função
calcular_calorias()
```
