```python
print("=== SELECIONE O QUE VOCÃŠ COMEU NO CAFÃ‰ DA MANHÃƒ ===")

# DicionÃ¡rio com os alimentos e seus valores nutricionais
menu = {
    "CafÃ© com aÃ§Ãºcar": {"calorias": 76, "carboidratos": 19, "proteinas": 0.2, "gorduras": 0},
    "Iogurte": {"calorias": 71, "carboidratos": 13, "proteinas": 1.7, "gorduras": 1.3},
    "Banana": {"calorias": 40, "carboidratos": 10.32, "proteinas": 0.52, "gorduras": 0.12},
    "PÃ£o de queijo (5 unidades)": {"calorias": 355, "carboidratos": 62.60, "proteinas": 11, "gorduras": 6.20}
}

# Exibir o menu
print("Menu do CafÃ© da ManhÃ£:")
for i, item in enumerate(menu.keys(), 1):
    print(f"{i}. {item}")
    
# Captura a escolha do usuÃ¡rio
escolhas = input("Digite os nÃºmeros dos itens que vocÃª comeu no cafÃ© da manhÃ£, separados por vÃ­rgula: ")
escolhas = [int(num.strip()) for num in escolhas.split(',') if num.strip().isdigit()]

# CÃ¡lculo dos totais usando list comprehension
calorias_totais, carboidratos_totais, proteinas_totais, gorduras_totais = [
    sum(menu[list(menu.keys())[i-1]][nutriente] for i in escolhas if 1 <= i <= len(menu))
    for nutriente in ["calorias", "carboidratos", "proteinas", "gorduras"]
]

# Exibir o total nutricional
print("\n=== TOTAL NUTRICIONAL DO CAFÃ‰ DA MANHÃƒ ===")
print(f"Calorias totais: {calorias_totais} kcal")
print(f"Carboidratos totais: {carboidratos_totais:.2f} g")
print(f"ProteÃ­nas totais: {proteinas_totais:.2f} g")
print(f"Gorduras totais: {gorduras_totais:.2f} g")
```

```python
def calcular_calorias():
    print("ğŸ‰ Bem-vindo Ã  Calculadora MÃ¡gica de Calorias DiÃ¡rias! ğŸ")
    
    # Solicitar dados do usuÃ¡rio
    print("\nVamos comeÃ§ar! Preencha os seguintes detalhes para descobrir quantas calorias vocÃª pode consumir por dia.")
    sexo = input("Qual o seu sexo? (homem/mulher): ").strip().lower()
    
    if sexo not in ['homem', 'mulher']:
        print("Ops! NÃ£o entendi... por favor, insira 'homem' ou 'mulher'. Vamos tentar de novo!")
        return
    
    peso = float(input("Qual o seu peso em quilos? âš–ï¸: "))
    altura = float(input("E qual a sua altura em centÃ­metros? ğŸ“: "))
    idade = int(input("Agora, me diga sua idade em anos! ğŸ‚: "))
    
    print("\nCalculando sua quantidade mÃ¡gica de calorias... ğŸ”®âœ¨")
    
    if sexo == "homem":
        # FÃ³rmula para homens
        calorias = (13.75 * peso) + (5 * altura) - (6.76 * idade) + 66.5
        print(f"\nğŸ’ª Uau! Para um homem com essas caracterÃ­sticas, vocÃª pode consumir atÃ© {calorias:.2f} calorias por dia! ğŸ¥³")
    elif sexo == "mulher":
        # FÃ³rmula para mulheres
        calorias = (9.56 * peso) + (1.85 * altura) - (4.68 * idade) + 665
        print(f"\nğŸ’ƒ Para uma mulher com essas caracterÃ­sticas, vocÃª pode consumir atÃ© {calorias:.2f} calorias por dia! ğŸ’–")
    
    print("\nAgora que vocÃª sabe, use essa informaÃ§Ã£o com sabedoria! ğŸ˜Œâœ¨")
    print("Lembre-se: o equilÃ­brio Ã© a chave para uma vida saudÃ¡vel! ğŸ§˜â€â™€ï¸ğŸ")

# Chamar a funÃ§Ã£o
calcular_calorias()
```
